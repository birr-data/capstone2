```python
def viewer(cat):
    df_win = df.loc[df['cat_descript'] == cat, ['cat_descript',  'sub_descript',  'sec_descript',  'subsec_descript',  'mod_descript']]
    print('CATEGORY: ', cat)
    print("")
    print('sub_descript:')
    print(df_win.sub_descript.value_counts())
    print('sec_descript:')
    print(df_win.sec_descript.value_counts())
    print('subsec_descript:')
    print(df_win.subsec_descript.value_counts())
    print('mod_descript:')
    print(df_win.mod_descript.value_counts())     
    
    return df_win
```
